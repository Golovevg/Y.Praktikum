```python
#Получился большой разброс значений, в том видны числа 26,27,32, которые вероятно являются ощибками записи 
#соответствующих значений 2.6,2.7,3.2. И цифра 100, которая судя по цене является площадью. Цифры меньше 2...
#Поэтому категорирую данные по высоте потолка.

#Пишу функцию категорирования 
def categorize_ceiling(ceiling_height):
    if ceiling_height < 2:
        return 'Ошибка записи по нижнему значению'
    if 2.4 > ceiling_height > 2:
        return 'От 2 до 2.4'
    if 2.7 > ceiling_height > 2.5:  
        return 'От 2.5 до 2.7'
    if 3.7 > ceiling_height > 2.8:
        return 'От 2.8 до 3.7'
    if 5 > ceiling_height > 3.8:
        return 'От 3.8 до 5'
    if ceiling_height > 5:
        return 'Ошибка записи по верхнему значению'

#Применяю функцию к столбцу с высотой потолков
df['categorize_ceiling'] = df.ceiling_height.apply(categorize_ceiling)

#Вывожу сводную таблицу по медиане (так как потолки обычно стандартные), добавлю среднее для визуального сравнения
#и посчитаю количество
df_ceiling = df.pivot_table(index = 'categorize_ceiling', values = ['last_price', 'total_area']
,aggfunc = ['median', 'count', 'mean'])
df_ceiling.head(10)
```

```python
#Большой разброс. Категорирую кухни:

def categorize_kitchen(kitchen):
    if kitchen < 2:
        return 'Ошибка записи по нижнему значению'
    if 4 > kitchen > 2:
        return '1'
    if 6 > kitchen > 4.1:  
        return '2'
    if 8 > kitchen > 6.1:
        return '3'
    if 12 > kitchen > 8.1:
        return '4'
    if 20 > kitchen > 12.1:
        return '5'
    if 40 > kitchen > 20.1:
        return '6'
    if 60 > kitchen > 40.1:
        return '7'    
    if 80 > kitchen > 60.1:
        return '8'
    if      kitchen > 80:
        return 'Зачем'

#Применяю функцию к столбцу с площадью кухни
df['categorize_kitchen'] = df.kitchen_area.apply(categorize_kitchen)

#Вывожу сводную таблицу по медиане и посчитаю количество
df_kitchen = df.pivot_table(index = 'categorize_kitchen', values = ['kitchen_area', 'total_area']
,aggfunc = ['count', 'median'])
df_kitchen.head(10)
```

```python
#Функция для замены значений, исходя из данных сводной таблицы
def kitchen_fill(i):
    if i < 28:
        return 3
    if 43 > i >= 28:
        return 5
    if 46 > i >= 43:
        return 7
    if 52 > i >= 46:
        return 10
    if 75 > i >= 52:
        return 14
    if 101 > i >= 75:
        return 26
    if 140 > i >= 101:
        return 45
    if 174 > i >= 140:
        return 65
    if i >= 174:
        return 125
#Создал столбец со значениями
df['kitchen_fill'] = df['total_area'].apply(kitchen_fill)


#Заменил пропуски значениями из столбца 'kitchen_fill'
df['kitchen_area'] = df.kitchen_area.combine_first(df.kitchen_fill)

#Удалил лишнее
del df['kitchen_fill']
del df['categorize_kitchen']
```
